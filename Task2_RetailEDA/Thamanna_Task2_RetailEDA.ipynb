```
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

df.head()
```

```
df.groupby('sex')['total_bill'].mean()
```

```
df.groupby('sex')['tip'].mean()
```

```
import matplotlib.pyplot as plt

df.groupby('day')['total_bill'].sum().plot(kind='bar')
plt.title("Total Sales by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill")
plt.show()
```

```
df.groupby('time')['total_bill'].sum().plot(kind='bar')
plt.title("Sales by Time (Lunch vs Dinner)")
plt.xlabel("Time")
plt.ylabel("Total Bill")
plt.show()
```

```
df.groupby('day')['tip'].mean().plot(kind='bar')
plt.title("Average Tip by Day")
plt.xlabel("Day")
plt.ylabel("Tip")
plt.show()
```

## 📊 Final Conclusion

- Customers spend an average amount per bill around the dataset mean.
- Certain days generate higher revenue compared to others.
- Dinner time shows higher sales than lunch.
- Male and female customers show different spending and tipping behavior.
- Overall, the dataset provides useful insights into customer purchasing patterns and business performance.
